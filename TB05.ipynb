とすればよい。左から順番に、赤、青、赤、青となっており、番号を1,2,3,4とする。このとき、左の赤(軌道1)は
- x方向に-1ずれた青色（軌道4）とy方向に+1ずれた青色(軌道2)および内部の青色（軌道2）

と接続しており、内部の青（軌道2）は

- 左端の赤色（軌道1）と内部の赤色(軌道3)と、y方向に-1ずれた赤色（軌道1）

、内部の赤（軌道3）は

- 内部の青色（軌道2）と右端の青色（軌道4）と、y方向に-1ずれた青色(軌道4)

、左端の青色(軌道4)は

- 内部の赤色(軌道3)とx方向に+1ずれた赤色（軌道1）およびy方向に+1ずれた赤色(軌道3)

と接続している。つまり、
- (1,ix,iy):(4,ix+1,iy),(2,ix,iy+1),(2,ix,iy)
- (2,ix,iy):(1,ix,iy),(3,ix,iy),(1,ix,iy-1)
- (3,ix,iy):(2,ix,iy),(4,ix,iy),(4,ix,iy-1)
- (4,ix,iy):(3,ix,iy),(1,ix+1,iy),(3,ix,iy+1)

というラベルを割り当てることができる。

この接続を反映させたハミルトニアンを作ると

```julia
function calc_HGrapheneA(Nx,Ny,μ)
    N = Nx*Ny*4
    mat_Htb = zeros(Float64,N,N)   
    mat_Htb += (-μ)*eye(N)
    t = 1.0
    for ix =1:Nx
        for iy=1:Ny
            for dx in -1:1
                for dy in -1:1
                    jx = ix + dx
                    jx += ifelse(jx > Nx,-Nx,0)
                    jx += ifelse(jx < 1,Nx,0)                    
                    jy = iy + dy
                    jy += ifelse(jy > Ny,-Ny,0)
                    jy += ifelse(jy < 1,Ny,0)  
                    for a=1:4
                        for b=1:4
                            hop = false
                            ii = ((iy-1)*Nx+ix-1)*4+a
                            if a == 1 
                                if dx==-1 && dy==0 && b==4
                                    hop = true
                                elseif dx==0 && dy==1 && b==2
                                    hop = true
                                elseif dx==0 && dy==0 && b==2
                                    hop = true
                                end                                                    
                            elseif a ==2
                                if dx==0 && dy==0 && b==1
                                    hop = true
                                elseif dx==0 && dy==0 && b==3
                                    hop = true
                                elseif dx==0 && dy==-1 && b==1
                                    hop = true
                                end
                            elseif a ==3
                                if dx==0 && dy==0 && b==2
                                    hop = true
                                elseif dx==0 && dy==0 && b==4
                                    hop = true
                                elseif dx==0 && dy==-1 && b==4
                                    hop = true
                                end                                
                            elseif a ==4
                                if dx==0 && dy==0 && b==3
                                    hop = true
                                elseif dx==1 && dy==0 && b==1
                                    hop = true
                                elseif dx==0 && dy==1 && b==3
                                    hop = true
                                end                                
                            end
                
                            if hop && 1 <= jx <= Nx && 1 <= jy <= Ny
                                jj = ((jy-1)*Nx+jx-1)*4+b                                                                                     
                                mat_Htb[ii,jj] = t
                            end
                        end
                    end
                end
            end
        end
    end
            
                                            
    return mat_Htb
end
```

となる。これを対角化しヒストグラムを見てみると、

```julia
Nx = 20
Ny = 40
μ = 0
mat_H = calc_HGrapheneA(Nx,Ny,μ)
#println(mat_H)
energy,mat_v = eig(mat_H)
histogram(energy,bins=100)
```

となり、ジグザグ型と同じ状態密度が得られる。この模型はx方向にもy方向にも周期的境界条件が課されている。次に、x方向にフーリエ変換してバンドをみてみよう。ハミルトニアンは

```julia
function calc_HGrapheneAkx(kx,Ny,μ)
    N = Ny*4
    mat_Htb = zeros(Complex{Float64},N,N)   
    mat_Htb += (-μ)*eye(N)
    t0 = 1.0
    #for ix =1:Nx
        for iy=1:Ny
        #    for dx in -1:1
                for dy in -1:1
                    #jx = ix + dx
                    #jx += ifelse(jx > Nx,-Nx,0)
                    #jx += ifelse(jx < 1,Nx,0)                    
                    jy = iy + dy
                    jy += ifelse(jy > Ny,-Ny,0)
                    jy += ifelse(jy < 1,Ny,0)  
                    t = 0.0
                    for a=1:4
                        for b=1:4
                            hop = false
                            t = 0.0
                            ii = (iy-1)*4+a
                            
                            if a == 1 
                                if dy==0 && b==4
                                    hop = true
                                    t += t0*exp(-im*kx)
                                elseif dy==1 && b==2
                                    hop = true
                                    t += t0
                                elseif dy==0 && b==2
                                    hop = true
                                    t += t0
                                end                                                    
                            elseif a ==2
                                if dy==0 && b==1
                                    hop = true
                                    t += t0
                                elseif dy==0 && b==3
                                    hop = true
                                    t += t0
                                elseif dy==-1 && b==1
                                    hop = true
                                    t += t0
                                end
                            elseif a ==3
                                if dy==0 && b==2
                                    hop = true
                                    t += t0
                                elseif dy==0 && b==4
                                    hop = true
                                    t += t0
                                elseif dy==-1 && b==4
                                    hop = true
                                    t += t0
                                end                                
                            elseif a ==4
                                if dy==0 && b==3
                                    hop = true
                                    t += t0
                                elseif dy==0 && b==1
                                    hop = true
                                    t += t0*exp(im*kx)
                                elseif dy==1 && b==3
                                    hop = true
                                    t += t0
                                end                                
                            end
                
                            if hop && 1 <= jy <= Ny
                                jj = (jy-1)*4+b                                                                                     
                                mat_Htb[ii,jj] = t
                            end
                        end
                    end
                end
    #        end
        end
    #end
            
                                            
    return mat_Htb
end
```

となり、kxを横軸として固有値をプロットすると

```julia
μ=0.0
Ny = 50
nkx = 100
vkx = linspace(-π,π,nkx)
ep = zeros(Float64,nkx,Ny*4)
cnt = 0
for kx in vkx    
    cnt += 1
    mat_H = calc_HGrapheneAkx(kx,Ny,μ)
    energy,mat_v = eig(mat_H)
    
    for i=1:Ny*4
        #println(energy[i])
        ep[cnt,i] = energy[i]
    end
end
plot(vkx,ep)
plot!(legend=false)
```
